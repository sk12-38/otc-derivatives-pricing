# I. Pricing, Interest Rate
이번 세션에서는 파생상품을 비롯한 금융상품의 평가와 이자율에 대한 기초 내용을 학습합니다.<br>

## 1. 이자율 기초

이자율은 채권에서 원금에 더해지는 이자의 비율을 의미합니다.<br>
하지만 채권의 종류가 다양하기에 같은 이자율을 가진 채권이어도 다른 가격을 가질 수 있습니다.<br>
따라서 금융상품을 평가하기 위해서는 이자율의 종류에 대해 공부할 필요가 있습니다.<br>
기본적으로 이자율은 이자에도 이자가 붙는지 여부에 따라 복리/단리로 나누어집니다.<br>
그 이외의 이자율의 분류 방법은 다음과 같습니다.


```mermaid
graph LR
    A[이자율] --> B[지급 횟수에 따른 분류]
    A --> C[이표 지급에 따른 분류]
    
    B --> D[이산횟수 지급]
    B --> E[연속지급]
    
    C --> F[이표채 금리]
    C --> G[무이표채 금리]


시장에 존재하는 주요 채권과 해당 채권의 현금흐름은 다음과 같습니다.<br>
원금 : $N$<br>
이자율 : $r$<br>
만기 : $T$<br>
이표채의 1년당 이자 지급 횟수 : $m$<br>
이표채 : 이자주기마다 $N \times {r \over m}$, 만기에 $N + N \times {r \over m}$<br>
할인채 : 만기에 $N$ (발행시 $N \over (1+r)^T $ 의 가격으로 판매)<br>
복리채 : 만기에 $N \times (1+r)^T$ <br>

아래 채권은 평가를 위해 사용하는 가상의 채권입니다.
1년당 이자 지급 횟수가 존재하는 복리채 : 만기에 $N \times (1+{r \over m})^{mT}$<br>
연속복리채 : 만기에 $N \times e^{rT}$<br>
연속복리할인채 : 만기에 $N$ (발행시 $N \over {e^{rT}} $ 의 가격으로 판매)<br>



## 2. 평가의 기초
### 2-1. 무차익거래
금융상품의 평가란, 장내시장에서 매매되지 않는 상품의 적절한 가치를 금융공학적 기술을 이용해서 측정하는 과정을 말합니다.<br>
평가를 진행할 때, 가장 중요한 법칙은 무차익거래의 법칙입니다.<br>
무차익거래의 법칙이란, 동일한 상품은 반드시 한 개의 가격만을 가져야 한다는 법칙입니다.<br>
여기서 ‘동일한 상품’이란 모든 상황에서 동일한 현금흐름을 발생시키는 상품을 뜻합니다.<br>
따라서 모든 상품의 평가는 다음과 같은 프로세스를 따릅니다.

$$
\boxed{\begin{array}{c}
\text{평가대상 상품의} \\
\text{상황별 현금흐름 분석}
\end{array}}
\;\;\;\;\Rightarrow\;\;\;\;
\boxed{\begin{array}{c}
\text{장내시장에 존재하는 상품들을} \\
\text{이용해 평가대상 상품의} \\
\text{현금흐름을 복제}
\end{array}}
\;\;\;\;\Rightarrow\;\;\;\;
\boxed{\begin{array}{c}
\text{해당 현금흐름을 복제하는} \\
\text{포트폴리오의 현재가치} \\
= \\
\text{평가대상 상품의 평가가격}
\end{array}}
$$

예시) 은행 A가 원금 10000원 이자율 5%의 2년 만기 복리채를 발행하고 있습니다. 이때 은행 A가 원금 10000원의 2년 만기 할인채를 발행하려고 합니다.<br>
은행 A가 2년 후에 1번의 현금흐름을 지급하는 두 채권은 평가 상으로 동일한 상품이며, 두 상품에 같은 금액을 투자한다면 만기에 받는 금액 또한 같아야 합니다.<br>
즉, 10000원을 투자하면 2년 후 11025원을 받는 복리채와 N원을 투자하면 2년 후 10000원을 받는 할인채에서 10000:11025 = N:10000의 비례식이 성립해야 한다는 것입니다. 따라서 해당 할인채의 가격은 9070.29원입니다.

### 2-2. 현재가치
현재가치란, 미래의 일정 금액을 현재가치로 환산한 것으로, 미래에 얻게 될 확실한 부의 가치를 현재의 가치로 환산한 값을 말합니다.<br>
그리고 미래의 현금흐름의 현재가치를 구하는 과정을 할인으로도 부르며, 할인에 쓰이는 이자율을 현물이자율(Spot Rate)로 부릅니다. <br>
금융공학에서는 확정되지 않은 미래의 현금흐름을 다루기에, 해당 현금흐름의 현재가를 구하는 할인 과정이 필수적입니다.<br>
금융공학에서 할인은, 일정한 주기마다 진행되는 것이 아니라 연속적으로 진행된다고 가정합니다. 따라서 위의 채권 중 연속복리할인채의 성격을 띠며,<br>
$T$년 후의 미래의 현금흐름 $N$의 현재가치 $PV$는 다음과 같습니다.<br>
$$PV = N \times {e^{-rT}}$$
이때 사용되는 이자율 $r$은 해당 현금흐름을 지급하는 주체의 부도위험에 따라 달라질 수 있으며, 부도위험이 없는 무위험 이자율은 이자율스왑에서 계산합니다.

### 2-3. 선도
선도(Forward)란 파생상품의 하나로, 미래의 일정 시기에 일정한 자산을 미리 정해진 가격에 거래하기로 약정하는 표준화되지 않은 계약을 뜻합니다.<br>
그리고 이때 정해진 가격을 선도가격(Forward Price)라고 합니다. 이러한 선도가격은 무차익거래 법칙을 통해 계산됩니다. <br>

예시1<br>
금 1kg의 현재 가격이 1억5천만원이고, 금 1kg을 보관하는 비용이 1년당 100만원이라고 가정합니다.<br>
만약 1년 후 금을 거래하는 선도가격이 1억6천만원이라면, 1년 후에 금을 매도하는 계약을 맺고,금 1kg을 구매, 100만원을 지불하고 금을 보관한 뒤 1년 뒤<br>
1억 6천만원에 매도한다면, 900만원의 무위험 수익을 얻을 수 있습니다. 이런 식의 무위험 수익을 없애기 위해서 선도가격은 1억 5100만원이 되어야 합니다.<br>

선도계약은 이자율에 대해서도 존재합니다. <br>

예시2<br>
은행 A에서 현재 1년 만기, 이자율 3%의 복리채와 2년 만기, 이자율 4%의 복리채를 동시에 발행하고 있다면,<br>
1년 뒤 은행 A에서 발행하는 1년 만기(2년 후 만기가 찾아오는) 복리채의 선도이자율은 다음과 같은 논리로 계산됩니다.<br>
Case A : 현재 원금 10000원의 2년 만기 복리채를 사는 경우 $\Rightarrow$ 2년 뒤 $ 10000 \times {(1.04)^2} = 10816$원 수취<br>
Case B : 현재 원금 10000원의 1년 만기 복리채를 산 뒤, 1년 뒤 받은 금액을 그대로 그날 발행된 1년 만기 복리채에 투자 $\Rightarrow$ $10000 \times (1.03) \times (1+r) $ 원 수취<br>
이때 두 Case 모두 2년 뒤 은행 A로부터 1번의 현금흐름을 받기에 두 Case에서 받는 금액은 동일해야 합니다. 따라서 선도이자율은 다음과 같습니다.<br>
$$r =  {10816 \over 10300 } - 1 \approx 5.0097 \% $$
